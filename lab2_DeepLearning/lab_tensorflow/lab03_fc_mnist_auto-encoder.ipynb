# TensorFlow 自动编码器(auto-encoder)

```python
# -*- coding: utf-8 -*-
#!/usr/bin/env python

from __future__ import print_function
```

导入MNIST数据集

```python
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets('./MNIST_data', one_hot=True)
print(
    '训练集大小: %s;\n验证集大小: %s;\n测试集大小: %s.' % \
    (mnist.train.num_examples, mnist.validation.num_examples, mnist.test.num_examples)
)
image_size = mnist.train.images[0].shape[0]
print('输入的大小：%s' % (image_size))
```

数据展示

```python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image

# plot original
original_image = mnist.train.images[1]
plt.imshow(np.reshape(a=original_image, newshape=(28, 28)) * 255)
```

```python
import tensorflow as tf

# hyper-parameter
training_epoch = 100
learning_rate = 1e-3
batch_size = 100

graph = tf.Graph()
with graph.as_default():
    with tf.name_scope('placehold'):
        x = tf.placeholder(dtype=tf.float32, shape=[None, image_size], name='x')


    def linear(input_data, name=None, target=None, activation=tf.nn.relu):
        original = int(list(input_data.shape)[-1])
        with tf.name_scope('layer_%s' % (name)):
            weight_a = tf.Variable(tf.random_normal(shape=[original, target], mean=.0, stddev=.1),
                                   name='layer_%s_a' % (name))
            weight_b = tf.Variable(tf.zeros(shape=[target]), name='layer_%s_b' % (name))
        return activation(tf.matmul(input_data, weight_a) + weight_b)


    # encoder
    encoder = linear(input_data=x, name='1', target=256)
    encoder = tf.nn.dropout(encoder, keep_prob=0.2)
    encoder = linear(input_data=encoder, name='2', target=128)
    encoder = tf.nn.dropout(encoder, keep_prob=0.2)
    encoder = linear(input_data=encoder, name='3', target=5)
    # decoder
    decoder = linear(input_data=encoder, name='4', target=128)
    decoder = tf.nn.dropout(decoder, keep_prob=0.2)
    decoder = linear(input_data=decoder, name='5', target=256)
    decoder = tf.nn.dropout(decoder, keep_prob=0.2)
    decoder = linear(input_data=decoder, name='6', target=image_size, activation=tf.nn.sigmoid)
    x_ = decoder

    with tf.name_scope('loss_op'):
        loss_op = tf.losses.mean_squared_error(x, x_)
    with tf.name_scope('train_op'):
        train_op = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss_op)
```

```python
with tf.Session(graph=graph) as session:
    session.run(tf.global_variables_initializer())

    # visualization
    img = np.reshape(a=original_image, newshape=(28, 28))
    blank = np.ones(shape=(28, 1))
    img = np.concatenate((img, blank), axis=-1)

    batch_episode = len(mnist.train.images) // batch_size
    for i in xrange(1, 1 + training_epoch):
        cost = 0.0
        for _ in xrange(batch_episode):
            batch_x, _ = mnist.train.next_batch(batch_size)
            feed = {x: batch_x}
            _, batch_cost = session.run([train_op, loss_op], feed_dict=feed)
            cost += batch_cost
        if i % 10 == 0:
            print('Epoch %s / %s, training cost: %s' % (i, training_epoch, cost / batch_episode))
            img_pred = x_.eval({x: original_image[None, ...]})[0]
            img = np.concatenate((img, blank, np.reshape(a=img_pred, newshape=(28, 28))), axis=-1)
    plt.imshow(img * 255)
    
    # 降维使用
    features, y_pred = session.run([encoder, decoder], feed_dict={
        x: mnist.test.images
    })
```

```python
Image.fromarray(img * 255).convert('L')
```

auto-encoder的思想是：编码+解码<br>
这现今的很多模型设计都有着非常重要的影响。因为这一过程可以看做：从数据中获得特征（编码），从特征复原数据（解码）。目前很多sequence-t0-sequence模型都是采用的encoder+decoder模式<br>
<br>
在本例中，我们可以利用encoder的输出--32个特征--作为MNIST数据的特征。将这些特征送到t-SNE中，降维到2维，解释MNIST数据的分布。

```python
# t-SNE
from sklearn.manifold import TSNE

features = (np.max(features) - features) / (np.max(features) - np.min(features))  # normalization
print('降维前，特征值的大小为：', features.shape)
lower_val = TSNE(n_components=2).fit_transform(features)  # dimensionality reduction
print('降维后，特征值的大小为：', lower_val.shape)
```

```python
plt.scatter(lower_val[:, 0], lower_val[:, 1], c=np.argmax(mnist.test.labels, axis=-1))
plt.colorbar()
plt.plot()
```

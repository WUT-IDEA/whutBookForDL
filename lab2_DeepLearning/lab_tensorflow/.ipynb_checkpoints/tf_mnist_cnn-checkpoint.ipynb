```python
# -*- coding: utf-8 -*-
# !/usr/bin/env python
'''
@author: Yang
@time: 17-12-23 下午9:40
'''

from __future__ import print_function

import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets('MNIST_data', one_hot=True)
print(
    'train data: %s;\nvalidation data: %s;\ntest data: %s.' % \
    (mnist.train.num_examples, mnist.validation.num_examples, mnist.test.num_examples)
)

image_size = 28
print('image size of MNIST is %s × %s' % (image_size, image_size))

import numpy as np
from PIL import Image

trainingEpochs = 100
n_class = 10

graph = tf.Graph()
with graph.as_default():
    with tf.name_scope('placehold'):
        # placehold - define variable
        x_data = tf.placeholder(dtype=tf.float32, shape=[None, image_size ** 2], name='x')
        # -1 for image number; 28,28 for image size; 1 for color channel
        X = tf.reshape(x_data, shape=(-1, image_size, image_size, 1))
        y_data = tf.placeholder(dtype=tf.int32, shape=[None, n_class], name='y')


    def fc_layer(input_data, units=None, activation=tf.nn.relu, name=None):
        return tf.layers.dense(inputs=input_data,
                               units=units,
                               activation=activation,
                               name='layer_%s' % (name))


    def conv_layer(input_data, filters=16, kernel_size=(2, 2),
                   activation=tf.nn.relu, padding='valid', name=None):
        with tf.name_scope('layer_%s' % (name)):
            input_data = tf.layers.conv2d(inputs=input_data,
                                          filters=filters,
                                          kernel_size=kernel_size,
                                          strides=(1, 1),
                                          activation=activation,
                                          padding=padding)
            input_data = tf.layers.max_pooling2d(input_data, pool_size=(2, 2), strides=(1, 1), padding=padding)
            return input_data


    conv1 = conv_layer(input_data=X, name='cnn1')
    conv2 = conv_layer(input_data=conv1, name='cnn2')
    fc1 = tf.contrib.layers.flatten(conv2)
    fc2 = fc_layer(input_data=fc1, units=1024, name='fc2')
    fc3 = tf.layers.dropout(fc2, rate=0.75)
    y_pred = fc_layer(input_data=fc3, units=n_class, name='fc3')

    with tf.name_scope('loss_func'):
        loss_op = tf.reduce_mean(
            tf.nn.softmax_cross_entropy_with_logits(
                logits=y_pred, labels=y_data))

    with tf.name_scope('optimizer'):
        optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss_op)

    with tf.name_scope('accuracy'):
        acc_op = tf.metrics.accuracy(labels=y_data, predictions=y_pred)

with tf.Session(graph=graph) as session:
    session.run(tf.global_variables_initializer())

    for i in xrange(trainingEpochs):

        cost = 0.0
        btach_size = 500
        total_batch = mnist.train.num_examples // btach_size
        for j in xrange(total_batch):
            batch_x, batch_y = mnist.train.next_batch(total_batch)
            _, batch_cost = session.run([optimizer, loss_op], feed_dict={
                x_data: batch_x,
                y_data: batch_y
            })
            cost += batch_cost
        cost /= total_batch
        print('Epoch %s / %s, training cost: %s' % (i + 1, trainingEpochs, cost))

    # test
    feed = {
        x_data: mnist.test.images,
        y_data: mnist.test.labels
    }
    _, acc = session.run([optimizer, acc_op], feed_dict=feed)
    print('Accuracy: %s' % (trainingEpochs, acc))
```
